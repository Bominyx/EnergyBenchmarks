```python
from datasets import load_dataset
from transformers import AutoTokenizer
import matplotlib.pyplot as plt

models = ["openai-community/gpt2", "meta-llama/Llama-3.2-1B", "Qwen/Qwen3-0.6B-Base"]
for model in models:
    ds = load_dataset("agentlans/high-quality-english-sentences", split="test")


    tokenizer = AutoTokenizer.from_pretrained(model)

    token_lengths = [
        len(tokenizer.encode(example["text"], add_special_tokens=False))
        for example in ds
    ]

    plt.figure(figsize=(10, 6))
    plt.hist(token_lengths, bins=50, color="#4C72B0", edgecolor="black", alpha=0.8)

    #plt.title(f"Verteilung der Tokenanzahl ({model}-Tokenisierung)", fontsize=14)
    print(f"Verteilung der Tokenanzahl ({model}-Tokenisierung)")
    plt.xlabel("Tokenanzahl", fontsize=12)
    plt.ylabel("Anzahl Datenpunkte", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    print(f"Höchste Tokenanzahl: {max(token_lengths)}")
    plt.show()

```

```python
from datasets import load_dataset
from transformers import AutoTokenizer
import matplotlib.pyplot as plt

models = [
    "openai-community/gpt2-xl",
    "meta-llama/Llama-3.2-1B",
    "Qwen/Qwen3-0.6B-Base"
]

colors = ["#4C72B0", "#55A868", "#C44E52"]  # drei Farben für die Modelle

plt.figure(figsize=(10, 6))

for model, color in zip(models, colors):
    ds = load_dataset("agentlans/high-quality-english-sentences", split="test")
    tokenizer = AutoTokenizer.from_pretrained(model)

    token_lengths = [
        len(tokenizer.encode(example["text"], add_special_tokens=False))
        for example in ds
    ]

    plt.hist(
        token_lengths,
        bins=50,
        color=color,
        edgecolor="black",
        #alpha=0.5,
        label=model
    )

    print(f"{model} – höchste Tokenanzahl: {max(token_lengths)}")

#plt.title("Verteilung der Tokenanzahl für verschiedene Tokenisierungen", fontsize=14)
print("Verteilung der Tokenanzahl für verschiedene Tokenisierungen")
plt.xlabel("Tokenanzahl", fontsize=12)
plt.ylabel("Anzahl Datenpunkte", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
